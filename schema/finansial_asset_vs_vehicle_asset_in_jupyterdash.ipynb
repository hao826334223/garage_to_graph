```python
app = JupyterDash(__name__)
app.layout = html.Div([
    html.H1("End to End Cross-Asset Graph"),
    dcc.Graph(id='graph'),
    html.Label([
        "Make_Model",
        dcc.Dropdown(
            id='make_model-dropdown', clearable=False,
            value=default_make_model, options=[
                {'label': c, 'value': c}
                for c in make_model_list
            ]),
    ]),
    html.Label([
        "Location",
        dcc.Dropdown(
            id='location-dropdown', clearable=False,
            value=locations_list[0], options=[
                {'label': c, 'value': c}
                for c in locations_list
            ]),
    ]),
    html.Label([
        "Financial Asset",
        dcc.Dropdown(
            id='fx-dropdown', clearable=False,
            value=list(fx_dict.keys())[0], options=[
                {'label': c, 'value': c}
                for c in list(fx_dict.keys())
            ]),
    ]),
])

# Define callback to update graph
@app.callback(
    Output('graph', 'figure'),
    [Input("make_model-dropdown", "value"), Input("location-dropdown", "value"), Input("fx-dropdown", "value")]
)
def update_figure(make_model, location, fx_raw):
    fx = fx_dict[fx_raw]
    fx_inv = fx_inv_dict[fx_raw]
    make = make_model.split('_')[0]
    model = make_model.split('_')[1]
    if len(make_model.split('_')) >= 2:
        model = '_'.join(make_model.split('_')[1:])

    
    fig = make_subplots(specs=[[{"secondary_y": True}]])
    
    if location == 'All':
        value_key = 'value'
    elif location == 'CA':
        value_key = 'value_ca'
    elif location == 'US':
        value_key = 'value_us'
        
    xy_day_car = [[y[0][list(y[0].keys())[0]], y[1]] for y in [[l[list(l.keys())[2]], l[value_key]] for l in g.V().has(
        'make', make).has('model', model).outE().has('time_type', 'day').elementMap().toList()]]
    df_day_car = pd.DataFrame(xy_day_car, columns = ['day','Vehicle_Asset'])
    df_day_car['Vehicle_Asset'] = df_day_car['Vehicle_Asset']/1.27
    df_day_car = df_day_car.groupby('day').mean().sort_values(by=['day'])
#     df_day_car.index.name = None
    df_day_car.index = [str(l)[0:4] + '-' + str(l)[4:6] + '-' + str(l)[6:8] for l in df_day_car.index.values]
    df_day_car.index = pd.to_datetime(df_day_car.index)
    df_day_car.index.names = ['date']
    df_day_car['Vehicle_Asset_5_Moving_Average'] = df_day_car['Vehicle_Asset'].rolling(5).mean()
    df_day_car['Vehicle_Asset_10_Moving_Average'] = df_day_car['Vehicle_Asset'].rolling(10).mean()
    df_day_car.reset_index(inplace= True)

#     xy_month_car = [[y[0][list(y[0].keys())[0]], y[1]] for y in [[l[list(l.keys())[2]], l['value']] for l in g.V().has(
#         'make', make).has('model', model).outE().has('time_type', 'month').elementMap().toList()]]
#     df_month_car = pd.DataFrame(xy_month_car, columns = ['month','Cars_Baseline_Price'])
#     df_month_car = df_month_car.groupby('month').mean().sort_values(by=['month'])
# #     df_month_car.index.name = None
#     df_month_car.index = [str(l)[0:4] + '-' + str(l)[4:6] for l in df_month_car.index.values]
#     df_month_car.index = pd.to_datetime(df_month_car.index)
#     df_month_car.index.names = ['date']
#     df_month_car['Cars_Baseline_Price_5_Moving_Average'] = df_month_car['Cars_Baseline_Price'].rolling(5).mean()
#     df_month_car['Cars_Baseline_Price_10_Moving_Average'] = df_month_car['Cars_Baseline_Price'].rolling(10).mean()
#     df_month_car.reset_index(inplace= True)
    df_day_car_tmp = df_day_car.copy()
    df_day_car_tmp['date'] = df_day_car_tmp['date'].astype('datetime64[ns]')
    df_month_car = df_day_car_tmp.resample('M', on='date').mean()
    df_month_car.reset_index(inplace= True)
#     display(df_month_car.head())
    
    xy_year_car = [[y[0][list(y[0].keys())[0]], y[1]] for y in [[l[list(l.keys())[2]], l['value']] for l in g.V().has(
        'make', make).has('model', model).outE().has('time_type', 'year').elementMap().toList()]]
    df_year_car = pd.DataFrame(xy_year_car, columns = ['year','Vehicle_Asset'])
    df_year_car['Vehicle_Asset'] = df_year_car['Vehicle_Asset']/1.27
    df_year_car = df_year_car.groupby('year').mean().sort_values(by=['year'])
#     df_year_car.index.name = None
    df_year_car.index = [str(l) for l in df_year_car.index.values]
    df_year_car.index = pd.to_datetime(df_year_car.index)
    df_year_car.index.names = ['date']
    df_year_car['Vehicle_Asset_5_Moving_Average'] = df_year_car['Vehicle_Asset'].rolling(5).mean()
    df_year_car['Vehicle_Asset_10_Moving_Average'] = df_year_car['Vehicle_Asset'].rolling(10).mean()
    df_year_car.reset_index(inplace= True)

#     orders_df_usd = get_diff_after_n_bdays(orders_df, n_delay_days)

#     tmp_df0 = orders_df_usd.copy()
#     # display(tmp_df0.head(2))

#     tmp_df0.set_index('date', inplace=True)
#     tmp_df0.index = pd.to_datetime(tmp_df0.index)
#     tmp_df1 = tmp_df0.groupby(['date']).agg({'price_diff_5bdays': 'mean'}).reset_index()
#     # display(tmp_df1.head(2))

#     tmp_df1.set_index('date', inplace=True)
#     tmp_df1.index = pd.to_datetime(tmp_df1.index)
#     dfs = df_grouping(tmp_df1)

    xy_day_fx = [(y[0][list(y[0].keys())[0]], y[1]) for y in [[l[list(l.keys())[2]], l['value']] for l in g.V().hasLabel(
        fx).outE().has('time_type', 'day').elementMap().toList()]]
    xy_day_fx = sorted(xy_day_fx)
    
    if fx_inv:
        df_day_fx = pd.DataFrame(data=[1/l[1] for l in xy_day_fx],    # values
                              index=[l[0] for l in xy_day_fx],    # 1st column as index
                              columns=['Financial_Asset'])
    else:
        df_day_fx = pd.DataFrame(data=[l[1] for l in xy_day_fx],    # values
                              index=[l[0] for l in xy_day_fx],    # 1st column as index
                              columns=['Financial_Asset'])
    df_day_fx.index = [str(l)[0:4] + '-' + str(l)[4:6] + '-' + str(l)[6:8] for l in df_day_fx.index.values]
    df_day_fx.index = pd.to_datetime(df_day_fx.index)
    df_day_fx.index.names = ['date']
    df_day_fx['Financial_Asset_5_Moving_Average'] = df_day_fx['Financial_Asset'].rolling(5).mean()
    df_day_fx['Financial_Asset_10_Moving_Average'] = df_day_fx['Financial_Asset'].rolling(10).mean()
    df_day_fx.reset_index(inplace= True)
    df_day_fx.head(2)
    xy_month_fx = [(y[0][list(y[0].keys())[0]], y[1]) for y in [[l[list(l.keys())[2]], l['value']] for l in g.V().hasLabel(
        fx).outE().has('time_type', 'month').elementMap().toList()]]
    xy_month_fx = sorted(xy_month_fx)
    
    if fx_inv:
        df_month_fx = pd.DataFrame(data=[1/l[1] for l in xy_month_fx],    # values
                                index=[l[0] for l in xy_month_fx],    # 1st column as index
                                columns=['Financial_Asset'])
    else:
        df_month_fx = pd.DataFrame(data=[l[1] for l in xy_month_fx],    # values
                                index=[l[0] for l in xy_month_fx],    # 1st column as index
                                columns=['Financial_Asset'])
    df_month_fx.index = [str(l)[0:4] + '-' + str(l)[4:6] for l in df_month_fx.index.values]
    df_month_fx.index = pd.to_datetime(df_month_fx.index)
    df_month_fx.index.names = ['date']
    df_month_fx['Financial_Asset_5_Moving_Average'] = df_month_fx['Financial_Asset'].rolling(5).mean()
    df_month_fx['Financial_Asset_10_Moving_Average'] = df_month_fx['Financial_Asset'].rolling(10).mean()
    df_month_fx.reset_index(inplace= True)
    df_month_fx.head(2)
    xy_year_fx = [(y[0][list(y[0].keys())[0]], y[1]) for y in [[l[list(l.keys())[2]], l['value']] for l in g.V().hasLabel(
        fx).outE().has('time_type', 'year').elementMap().toList()]]
    xy_year_fx = sorted(xy_year_fx)

    
    if fx_inv:
        df_year_fx = pd.DataFrame(data=[l[1] for l in xy_year_fx],    # values
                                index=[l[0] for l in xy_year_fx],    # 1st column as index
                                columns=['Financial_Asset'])
    else:
        df_year_fx = pd.DataFrame(data=[l[1] for l in xy_year_fx],    # values
                                index=[l[0] for l in xy_year_fx],    # 1st column as index
                                columns=['Financial_Asset'])
    df_year_fx.index = [str(l) for l in df_year_fx.index.values]
    df_year_fx.index = pd.to_datetime(df_year_fx.index)
    df_year_fx.index.names = ['date']
    df_year_fx['Financial_Asset_5_Moving_Average'] = df_year_fx['Financial_Asset'].rolling(5).mean()
    df_year_fx['Financial_Asset_10_Moving_Average'] = df_year_fx['Financial_Asset'].rolling(10).mean()
    df_year_fx.reset_index(inplace= True)

#     display(df_year_car.head())
#     display(df_day_fx.head())
    
#     dfs_car = {'daily':df_day_car, 'monthly': df_month_car, 'yearly' :df_year_car}
#     dfs_fx = {'daily':df_day_fx, 'monthly': df_month_fx, 'yearly' :df_year_fx}
    
    dfs = {'daily':[df_day_car, df_day_fx], 'monthly':[df_day_car, df_month_fx], 'yearly' :[df_year_car, df_year_fx]}
    
#     dfs = {'daily':[df_day_car], 'monthly':[df_day_car], 'yearly' :[df_year_car]}
    
    
    # display(tmp_df1.head(2))

    # specify visibility for traces accross dataframes
    frames = len(dfs)  # number of dataframes organized in  dict
    # number of columns i df, minus 1 for Date
    columns = len(dfs['daily'][0].columns) + len(dfs['daily'][1].columns) - 1
    scenarios = [list(s) for s in [e == 1 for e in np.eye(frames)]]
    visibility = [list(np.repeat(e, columns)) for e in scenarios]

    # container for buttons
    buttons = []

    # iterate of dataframes in dfs:
    # - i is used to reference visibility attributes
    # - k is the name for each dataframe
    # - v is the dataframe itself
    for i, (k, vs) in enumerate(dfs.items()):
        for j, v in enumerate(vs):
#             print(vs)
            for c, column in enumerate(v.columns[1:]):
                if j == 0:
                    fig.add_scatter(name=column,
                                    x=v['date'],
                                    y=v[column],
                                    # 'daily' values are shown from the start
                                    visible=True if k == 'daily' else False,
                                    line=dict(color=['royalblue', "rgb(0, 102, 204)", "rgb(0, 204, 204)"][c], width=4 if c == 0 else 1)
                                    )
                else:
                    fig.add_scatter(name=column,
                                    x=v['date'],
                                    y=v[column],
                                    # 'daily' values are shown from the start
                                    visible=True if k == 'daily' else False,
                                    line=dict(color=['firebrick', "rgb(255, 0, 0)", "rgb(255, 128, 0)"][c], width=[4, 2, 1][c], dash='dash'),
                                    secondary_y=True,
                                    )

        # one button per dataframe to trigger the visibility
        # of all columns / traces for each dataframe
        button = dict(label=k,
                      method='restyle',
                      args=['visible', visibility[i]])
        buttons.append(button)
#     usd_cad_df_plot = usd_cad_df.copy()
#     usd_cad_df_plot = usd_cad_df[usd_cad_df.index > str(dfs['daily']['date'].values.min())]
#     usd_cad_df_plot['usd_cad_diff'] = usd_cad_df_plot['Open'].diff(periods=5) #.rolling(5).mean()
#     fig.add_scatter(x=usd_cad_df_plot.index, y=usd_cad_df_plot['Open'] * 500, name="usd_cad (* 500 for scale)")
#     fig.add_scatter(x=usd_cad_df_plot.index, y=usd_cad_df_plot['usd_cad_diff'] * 10000, name="usd_cad_diff (* 10000 for scale)")

    # include dropdown updatemenu in layout
    fig.update_layout(title=f"Financial Asset ({fx}) vs Vehicle Asset ({make_model} in USD)",
                      updatemenus=[dict(type="buttons",
                                        direction="right",
                                        x=0.8,
                                        y=1.1,
                                        buttons=buttons,
                                        xanchor="left",
                                        yanchor="bottom",
                                        #                                     align="right"
                                        )])
    # Set x-axis title
    fig.update_xaxes(title_text="Date")
    fig.update_yaxes(title_text="<b>Vehicle Asset</b> (USD)", secondary_y=False)
    fig.update_yaxes(title_text="<b>Financial Asset</b>", secondary_y=True)
    return fig
# app.run_server(port=8050)
```
